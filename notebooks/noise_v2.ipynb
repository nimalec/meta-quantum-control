```python
from metaqctrl.quantum.noise_models_v2 import NoisePSDModel, NoiseParameters, PSDToLindblad
import numpy as np 
psd_model = NoisePSDModel('one_over_f')
theta = NoiseParameters(alpha=1.0, A=1e3, omega_c=3.0)
HBAR = 1.054_571_817e-34
conv = PSDToLindblad(psd_model, g_energy_per_xi=HBAR/2)
ops, rates = conv.qubit_lindblad_ops(theta, T=50e-6, sequence='ramsey', omega0=2*np.pi*5e6)
print(rates)  # [Γ↓, Γ↑, γφ] in 1/s
```
